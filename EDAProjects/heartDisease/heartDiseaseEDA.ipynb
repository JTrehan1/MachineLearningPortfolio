```python
# import libraries
import codecademylib3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

# load data
heart = pd.read_csv('heart_disease.csv')

# Inspect the data 
print(heart.head())

# Boxplot to show distribution of thalch of those with heart disease or not

sns.boxplot(data=heart, x=heart.heart_disease, y=heart.thalach)
plt.xlabel("Heart Disease Presence")
plt.ylabel("Thalach Score")
plt.show()

# Saving the values of thalach with and without heart disease
hd = heart[heart["heart_disease"] =='presence' ]
thalach_hd = hd["thalach"]
no_hd = heart[heart["heart_disease"] =='absence' ]
thalach_no_hd = no_hd["thalach"]
print(thalach_hd)
print(thalach_no_hd)

mean_thalach_difference = thalach_no_hd.mean() - thalach_hd.mean()
print(mean_thalach_difference)

from scipy.stats import ttest_ind
# H0 The average thalach for a person with heart disease is equal to the average thalach for a person without heart disease
# H1 The average thalach for a person with heart disease is NOT equal to the average thalach for a person without heart disease

tstat, pval = ttest_ind(thalach_no_hd, thalach_hd)
print(pval) # 3.456964908430172e-14, reject H0

# Investigating the cholestrol in those with and without heart disease:
sns.boxplot(data=heart, x=heart.heart_disease, y=heart.chol)
plt.xlabel("Heart Disease Presence")
plt.ylabel("Chol Level")
plt.show()
plt.clf()

# Saving the values of chol with and without heart disease
hd = heart[heart["heart_disease"] =='presence' ]
chol_hd = hd["chol"]
no_hd = heart[heart["heart_disease"] =='absence' ]
chol_no_hd = no_hd["chol"]
print(chol_hd)
print(chol_no_hd)

mean_chol_difference = chol_no_hd.mean() - chol_hd.mean()
print(mean_chol_difference)

from scipy.stats import ttest_ind
# H0 The average Chol for a person with heart disease is equal to the average chol for a person without heart disease
# H1 The average Chol for a person with heart disease is NOT equal to the average chol for a person without heart disease

tstat, pval_chol = ttest_ind(chol_hd, chol_no_hd,)
print(pval_chol) # 0.13914167020436527 accept H0

# Chest pain and heart rate
sns.boxplot(data=heart, x=heart.cp, y=heart.thalach)
plt.xlabel("Heart Disease Presence")
plt.ylabel("Thalach Level")
plt.show()
plt.clf()

# Saving the values of Thalach for chest pain type
thalach_typical = heart.thalach[heart.cp =='typical angina']
thalach_asymptom = heart.thalach[heart.cp =='asymptomatic']
thalach_nonangin = heart.thalach[heart.cp =='non-anginal pain']
thalach_atypical = heart.thalach[heart.cp =='atypical angina']

from scipy.stats import f_oneway
Fstat, pval_cp = f_oneway(thalach_typical, thalach_asymptom, thalach_nonangin, thalach_atypical)
print(pval_cp)
# pval_cp = 1.9065505247705008e-10 reject H0 at 5% sl

# Heart disease and chest pain 
Xtab = pd.crosstab(heart.cp, heart.heart_disease)
print(Xtab)
from scipy.stats import chi2_contingency 
'''Null: There is NOT an association between chest pain type and whether or not someone is diagnosed with heart disease.
Alternative: There is an association between chest pain type and whether or not someone is diagnosed with heart disease.'''

chi2, pval_hd_cp, dof, expected = chi2_contingency(Xtab)
print(pval_hd_cp) # 1.2517106007837527e-17 reject H0
```
